Train Test split Rearranging Columns


```python
col = ['Season','Crop','State','Crop_Year','Fertilizer','Pesticide','Annual_Rainfall','Production','Yield']
df = df[col]
```

```python
x = df.drop('Yield',axis=1)
y = df['Yield']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_train
```

Converting Categorical to Numerical and Scaling the value


```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(drop='first')
scaler = StandardScaler()
```

```python
x_train.head(1)
```

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('OneHotEncoder',ohe,[0,1,2]),
        ('standrization',scaler,[3,4,5,6,7])
],
remainder = "passthrough"
)
        
```

```python
preprocessor
```

```python
x_train_dummy= preprocessor.fit_transform(x_train)
x_test_dummy = preprocessor.transform(x_test)
```

```python
x_train
```

```python
x_train_dummy
```

Training Models


```python
from  sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
models = {
    'lr':LinearRegression(),
     'lss':Lasso(),
     'rg':Ridge(),
      'Knr':KNeighborsRegressor(),
       'dtr':DecisionTreeRegressor()
}

for name,mod in models.items():
    mod.fit(x_train_dummy,y_train)
    y_pred = mod.predict(x_test_dummy)

    print(f"{name} MSE : {mean_absolute_error(y_test,y_pred)} Score {r2_score(y_test,y_pred)}")
```

Select Model


```python
dtr = DecisionTreeRegressor()
dtr.fit(x_train_dummy,y_train)
dtr.predict(x_test_dummy)
```

Predictive System


```python
def prediction(Season, Crop, State, Crop_Year, Fertilizer, Pesticide, Annual_Rainfall, Production):
    # Create an array of input features
    features = np.array([[Season, Crop, State, Crop_Year, Fertilizer, Pesticide, Annual_Rainfall, Production]])

    # Ensure proper transformation of features
    try:
        Transformed_features = preprocessor.transform(features)
    except Exception as e:
        print(f"Error during preprocessing: {e}")
        return None

    # Make prediction
    try:
        predicted_value = dtr.predict(Transformed_features)
    except Exception as e:
        print(f"Error during prediction: {e}")
        return None

    # Return the predicted value
    return predicted_value[0]  # No need to reshape if output is already a 1D array
```

```python
Season = 'Rabi'
Crop = 'Gram'
State = 'Himachal Pradesh'
Crop_Year = 2008
Fertilizer = 1.041331e+05	
Pesticide = 65.52
Annual_Rainfall = 1357.100000
Production = 290

result = prediction(Season,Crop,State,Crop_Year,Fertilizer,Pesticide,Annual_Rainfall,Production)
```
